```python
#Question 5.9
import imageio
import matplotlib.pyplot as plt
import numpy as np

image = imageio.imread('m101BW.jpg')

plt.subplot(1, 3, 1)
plt.imshow(image, cmap = 'jet')

plt.subplot(1,3,2)
plt.imshow(image, cmap = 'jet')
contour_levels = [50, 100, 150, 200] 
plt.contour(image, levels=contour_levels, colors='red')

filtered_image = np.where(image >= np.percentile(image, 90), image, 0)

plt.subplot(1, 3, 3)
plt.imshow(filtered_image, cmap = 'jet') 
```

```python
#Question 5.10
import numpy as np
import matplotlib.pyplot as plt

M = np.loadtxt('ngc6946.txt')
M.shape = (40, 50, 70)
S = M[:,-1,:]

z_profile = M[:, 25, 35]
print("Z-profile data : ", z_profile)

print(f'\nThe sum of the profile is {np.sum(z_profile)}')
print(f'The minimum of the profile is {np.min(z_profile)}')
print(f'The maximum value of the profile is {np.max(z_profile)}')
print(f'The mean of the profile is {np.mean(z_profile)}')
print(f'The standard deviation of the profile is {np.std(z_profile)}')

plt.hist(z_profile, bins = 20)
plt.text(0.065, 6, 'Sum = 0.89154')
plt.text(0.065, 5.7, 'Min Val = 0.00482')
plt.text(0.065, 5.4, 'Max Val = 0.0606')
plt.text(0.065, 5.1, 'Mean = 0.0222885')
plt.text(0.065, 4.8, 'Stdev = 0.012699434544498428')
```

```python
#Question 5.11
import numpy as np
import matplotlib.pyplot as plt

M = np.loadtxt('ngc6946.txt')
M.shape = (40, 50, 70)
plane = M[0, :, :]

m = np.mean(plane)
s = np.std(plane)

levels = [m, m + s, m + 2 * s]

plt.imshow(plane, cmap='viridis')
plt.colorbar()

plt.contour(plane, levels=levels, colors='white')

plt.title("First Plane of NGC 6946")
plt.xlabel("x-axis (columns)")
plt.ylabel("y-axis (rows)")
plt.show()
```

```python
#Question 5.12
from scipy.constants import c
class Galaxy:    
    """Class for redshift calculation of galaxies"""  
        
    caH_0 = 3968.5      # Calcium H line in rest frame
    lightspeed = c*3.6 #in km/h
    
    def __init__(self, name, calciumHline=caH_0):
        self.name = name
        self.caH = calciumHline
    def redshift(self):        
        z = (self.caH - self.caH_0)/self.caH
    def velocity(self, c = lightspeed):
        z = (self.caH - self.caH_0)/self.caH
        v = z*c
        return v

gal = Galaxy("NGC1832", 3994.7)
print(f'Velocity of {gal.name} galaxy is {gal.velocity()} km/h')
```
