```python
#Question 3.2
from math import *
from scipy.constants import G

masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']
velocity = []

for i in range(len(masses)) :
    v = sqrt((2*G*masses[i])/radii[i])
    print(f'The masses of the planets are {masses[i]:.2e}')
    velocity.append(v)

planetvel = dict(zip(names, velocity))
print(planetvel)
```

```python
#Question 3.7
from math import *
def angularsize(size, distance):
    A = 2 * atan(size/(2* distance))
    return A

B = float(input('The distance is '))
C = float(input('The size is '))

print(angularsize(C, B))
```

```python
#Question 3.10

from math import *
from scipy.constants import pi

def cylinder(height, radius) :
    A = 2*pi*(radius**2) + 2*pi*radius*height
    V = 2*height*(radius**2)
    return A, V

def ratio(height, radius):
    A = 2*pi*(radius**2) + 2*pi*radius*height
    V = 2*height*(radius**2)
    return A/V

B = float(input('The height is (in meters) '))
C = float(input('The radius is (in meters) '))

print(f'The area is {cylinder(B,C)[0]} and volume is {cylinder(B,C)[1]} of the cylinder ')
print(f'The ratio is {ratio(B, C)}')
```

```python
#Question 3.17
from math import *
import numpy as np

def stdev(mean, std, n):
    a = np.random.normal(mean, std, n)
    s = sqrt((1/len(a))*np.sum((a -(1/len(a))*np.sum(a))**2))
    return s 

e = float(input('The mean is '))
f = float(input('The standard deviation is '))
g = int(float(input('The number of values is ')))

x = np.random.normal(e, f, g)


print(np.std(x))
print(stdev(e, f, g))
    
```
