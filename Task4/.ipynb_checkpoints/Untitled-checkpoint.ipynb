```python
#Question 4.3 Done

import numpy as np
rng = np.random.default_rng()

X = input('Choose between R, P or S; Choosen is : ')

Y = ['R','P','S']
Z = rng.choice(Y)
print('The computer chooses :',Z)

if X == Z :
    print('Play Again')
elif X == 'R': 
    if Z == 'P':
        print('You lost')
    elif Z == 'S':
        print('You won')
elif X == 'S': 
    if Z == 'P':
        print('You won')
    elif Z == 'R':
        print('You lost')
elif X == 'P': 
    if Z == 'R':
        print('You won')
    elif Z == 'S':
        print('You lost')      
```

```python
#Question 4.4 Done
import numpy as np
import matplotlib.pyplot as plt
from math import *
import scipy

plt.rc('text', usetex=True)

X = np.random.normal(185, 8, 100000)
Mean = np.mean(X)
STD = np.std(X)
print(f'The mean is {Mean} and the standard deviation is {STD}')

plt.hist(X, 40, label = 'Random Distribution')
plt.axvline(Mean, color = 'red', label = 'Mean value is {}'.format(round(Mean)))
plt.axvspan(Mean-STD, Mean+STD, color='pink', alpha=0.5, label='Standard Deviation value is {}'.format(round(STD)))
plt.xlabel('Height')
plt.ylabel('Number')
plt.title('Height Distribution amoung a generated population')
plt.legend()

h = float(input('The height is : '))
print(f'The probability to have a height lower or equal to {h} is {scipy.stats.norm.cdf(h, Mean, STD)}')
print(f'The probability to have a height within the std range is {scipy.stats.norm.cdf(Mean + STD, Mean, STD)-scipy.stats.norm.cdf(Mean - STD, Mean, STD)}')
```

```python
#Question 4.6 Done

import numpy as np

def functioncalc(fun, x, a, b, n):
    y = eval(fun)
    S = 0 
    S = ((b - a)/n)*np.sum(y)
    return S


givenfun = input("Enter a function f(x) = ")
a = float(input('The lower boundary is : '))
b = float(input('The upper boundary is : '))
n = float(input('The precision is : '))
X = np.linspace(a, b, int(n))
print(f'The value of the function between {a} and {b} is : {functioncalc(givenfun, X, a, b, n)} ')      
    
```

```python
## Question 4.12 Done

from math import *
import numpy as np
from matplotlib import pyplot as plt

plt.rc('text', usetex=True)

angle_degree = 30
theta = radians(angle_degree)

rotor = cos(theta) + 1j*sin(theta)

x = np.array([1,3,3,1,1])
y = np.array([1,1,3,3,1])
plt.plot(x, y, color='red', marker='o', label = 'Initial Square')

xy = []
for x,y in zip(x, y):
    xy.append(complex(x,y))
print(xy)
xyrot = []
for i in xy:
    xyrot.append(rotor * i)
xyrot = np.array(xyrot)    

plt.plot(xyrot.real, xyrot.imag, color='blue', marker='o', label = 'Rotated Square')

plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Squares Printing')
plt.legend()
```

```python
#Question 4.14 Done
import numpy as np
from math import *

def funcfilter(func, sec):
    A = []
    B = np.asarray((sec))
    for i in B:
        if i > 0:
            A.append(func(i))
    return A
   
sq = np.array([-100, -10, 0, 10, 100])
result = funcfilter(np.log10, sq)
print("Result log10:", result)

sq2 = [100]
result2 = funcfilter(np.log10, sq2)
print("Result log10:", result2)
```

```python
#Question 4.16 Done

from astropy.cosmology import Planck18
from math import *
import scipy

z = float(input('The upper limit is : '))
H0 = Planck18.H0.value #km / (Mpc s)
H1 = H0
Om0 = Planck18.Om0 #no dimension
Ode0 = Planck18.Ode0 #no dimension

def F(x):
    H = H1*sqrt(Om0*((1+x)**3)+Ode0)
    T = 1/((1+x)*H)
    return T

tz = scipy.integrate.quad(F, 0, z)

Gyr_conversion = 977.79

print(tz[0]*Gyr_conversion)
```

```python
#Question 4.2 Done

import numpy as np 
import matplotlib.pyplot as plt
from scipy.stats import norm
plt.rc('text', usetex=True)

X = np.random.normal(0, 1, 100000)
Y = np.random.normal(2,1, 100000)


plt.hist(X, bins=30, alpha=0.7, label='X', density=True, color='blue', edgecolor='black', linewidth=1)
plt.hist(Y, bins=30, alpha=0.7, label='Y', density=True, color='purple', edgecolor='black', linewidth=1)
plt.title('1D Historigram and PDF')

#PDF
x_axis = np.linspace(-5, 7, 500)
plt.plot(x_axis, norm.pdf(x_axis, 0, 1), 'k:', lw=5,label='X(PDF)')
plt.plot(x_axis, norm.pdf(x_axis, 2, 1),'k--',lw=5,label='Y(PDF)')

#Arrows
plt.annotate(r'$\mathcal{N}(0, 1)$', xy=(-1, 0.15), xytext=(-3.5, 0.15),
             arrowprops=dict(facecolor='blue'), color='blue', fontsize=12)
plt.annotate(r'$\mathcal{N}(2, 1)$', xy=(3, 0.15), xytext=(4.5, 0.15),
             arrowprops=dict(facecolor='red'), color='red', fontsize=12)

plt.ylabel('Probability Density')
plt.xlabel('Values')
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.random.normal(0, 1, 100000)
Y = np.random.normal(2,1, 100000)

H, x_e, y_e = np.histogram2d(X, Y, bins=30, range=[[-5,5], [-2,6]])
H = H * (120 / H.max()) 
plt.imshow(H.T, extent=[*[-5,5], *[-2,6]], origin='lower', cmap='Blues')
plt.colorbar()

v = np.sort(H.flatten())[::-1]
levels = [v[np.searchsorted(np.cumsum(v), p * v.sum())] for p in [0.68, 0.95, 0.99]]

extent = [x_e[0], x_e[-1], y_e[0], y_e[-1]]
cs = plt.contour(H.T, levels=sorted(levels), colors='red', extent=extent)
plt.clabel(cs, fmt={levels[0]:'1$\sigma$', levels[1]:'2$\sigma$', levels[2]:'3$\sigma$'})

plt.title('2D Histogram + Contour')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(ls='--')
plt.show()
```
