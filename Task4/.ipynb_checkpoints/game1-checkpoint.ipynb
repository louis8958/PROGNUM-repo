```python
import numpy as np
rng = np.random.default_rng()

X = input('Choose between R, P or S; Choosen is : ')

Y = ['R','P','S']
Z = rng.choice(Y)
print('The computer chooses :',Z)

if X == Z :
    print('Play Again')
elif X == 'R': 
    if Z == 'P':
        print('lost')
    elif Z == 'S':
        print('won')
elif X == 'S': 
    if Z == 'P':
        print('won')
    elif Z == 'R':
        print('lost')
elif X == 'P': 
    if Z == 'R':
        print('won')
    elif Z == 'S':
        print('lost')  
```
